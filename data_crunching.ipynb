```python
alpha_values = list(np.unique(df[['alpha']].values))
beta_values = list(np.unique(df[['beta']].values))
skip_frames_values = list(np.unique(df[['skip_frames']].values))
conf_thresh_values = list(np.unique(df[['conf_thresh']].values))
iou_threshold_values = list(np.unique(df[['iou_threshold']].values))

deep_sort_values = list(np.unique(df[['deep_sort']].values))
tracker_values = list(np.unique(df[['tracker']].values))

# tracker_values = ['kalman']
```

```python
print(f'alpha_values = {alpha_values}')
print(f'beta_values = {beta_values}')
print(f'skip_frames_values = {skip_frames_values}')
print(f'conf_thresh_values = {conf_thresh_values}')
print(f'iou_threshold_values = {iou_threshold_values}')
print(f'deep_sort_values = {deep_sort_values}')
print(f'tracker_values = {tracker_values}')
```

```python
all_idf1 = []
for alpha, beta, skip_frames, conf_thresh, iou_threshold, deep_sort, tracker in product(alpha_values, beta_values, skip_frames_values, conf_thresh_values, iou_threshold_values, deep_sort_values, tracker_values):

    print(alpha, beta, skip_frames, conf_thresh, iou_threshold, deep_sort, tracker)

    idf1_mean = df[
    (df.alpha==alpha) & 
    (df.beta == beta) &
    (df.skip_frames == skip_frames) &
    (df.conf_thresh == conf_thresh) &
    (df.iou_threshold == iou_threshold) &
    (df.deep_sort == deep_sort) &
    (df.tracker == tracker)
    ].idf1.mean()

    print(idf1_mean)
    all_idf1.append([idf1_mean,[alpha, beta, skip_frames, conf_thresh, iou_threshold, deep_sort, tracker]])

print(f'biggest is {max(all_idf1, key = lambda x:x[0])}')
```
